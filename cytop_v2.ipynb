```python
# process with real parameters

import numpy as np
from matplotlib import pyplot as plt
import math
from scipy.special import erf

plt.rcParams["figure.figsize"] = [3, 3]
plt.rcParams["figure.autolayout"] = True

# pulse numbers

N=100

# material parameters cytop

k=0.12
ro=2030
Cp=861
Dth=k/(ro*Cp)

# material parameters sio2 glass

k2=1.38
ro2=2200
Cp2=703
Dth2=k2/(ro2*Cp2)


#laser parameters

f=10**7
Tp=1/f
P=0.15*0.37  #power arrives inT sample
Ep=P/f

NA=0.75
lamda=1030*10**-9
w=lamda/(math.pi*NA) #radius of the focus beam
wz=3*w

# accumulation parameter

Td=((2*w)**2)/(4*Dth)
Td2=((2*w)**2)/(4*Dth2)
Rt=Tp/Td
Rt2=Tp/Td2

# analytical calculation

A=0.005

T00=(A*Ep)/((math.pi**1.5)*ro*Cp*wz*w**2)
T002=(A*Ep)/((math.pi**1.5)*ro2*Cp2*wz*w**2)


Tave=T00*(2/Rt)*(1-(1/(1+N*Rt)**(1/2)))
Tavefinal=T00*(2/Rt)
rw=0.001
Taver=T00*((math.pi**0.5)/(rw*Rt))*(math.erf(rw)- math.erf((rw/(1+N*Rt)**(1/2))))
Taverfinal=T00*((math.pi**0.5)/(rw*Rt))*(math.erf(rw))

Nneed=(1/Rt)*((1/0.03)**2-1)


print("width=",2*w,"Dth=",Dth,"Td=",Td,"Rt=",Rt)
print("T00=",T00, "Tave=",Tave,"final Tave=",Tavefinal)
#print("pulse needed for ss=",Nneed, "time need", Nneed/f)
print("Taver=",Taver,"final Taver=",Taverfinal)

# process drawing

for n1 in range(N):
    t1 = np.linspace((n1)*Tp, (n1+1)*Tp, 100)

    def f1(t1):
        T1 = 0
        for i1 in range(n1):
             T1 += (T00/((1+((t1/Tp)-i1)*Rt)**(3/2))) * np.exp(-(rw**2)/(1+((t1/Tp)-i1)*Rt))
        return T1
    n1+=1
    plt.plot(t1, f1(t1), linewidth=1,color='red',label='CYTOP'if n1 == 1 else "") 
    
for n2 in range(N):
    t2 = np.linspace((n2)*Tp, (n2+1)*Tp, 100)

    def f2(t2):
        T2 = 0
        for i2 in range(n2):
             T2 += (T002/((1+((t2/Tp)-i2)*Rt2)**(3/2))) * np.exp(-(rw**2)/(1+((t2/Tp)-i2)*Rt2))
        return T2
    n2+=1
    plt.plot(t2, f2(t2), linewidth=1,color='blue',label='SiO2 glass'if n2 == 1 else "")     

#plt.xlim(0,1E-3)
#plt.ylim(0,3)
plt.xlabel('time(s)')
#plt.xlabel('reduced radius(r/w)')
plt.ylabel('Temperature(K)')
#plt.xticks(np.arange(0,0.0011,5E-4))
plt.legend(loc='upper left')
    
plt.show()  
```
